```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/HLOosvsPgIwt5dgOOh1RSg/survey-data-updated.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.columns.tolist())
```

```python
# Split 'LanguageWorkedWith' into a list
df['expLanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].str.split(';')

# Explode the list into separate rows
df_langhave_exploded = df.explode('LanguageHaveWorkedWith')
print ("list exploded")

df_langhave_exploded.to_csv("lang.csv", index=False)
print('saved')
```

```python
# Split 'Language want to work worked with' into a list
df['expLanguageWantToWorkWith'] = df['LanguageWantToWorkWith'].str.split(';')

# Explode the list into separate rows
df_Langwant_exploded = df.explode('LanguageWantToWorkWith')
print ("list exploded")
df_Langwant_exploded.to_csv("lang.csv", index=False)
print('saved')
```
